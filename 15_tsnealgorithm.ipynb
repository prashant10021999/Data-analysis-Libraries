```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
data,labels = make_blobs(n_samples=300, centers=3,random_state=42)
```

```python
data_standardized = StandardScaler().fit_transform(data)
```

```python
tsne = TSNE(n_components=2,random_state=42)
data_tsne = tsne.fit_transform(data_standardized)
```

```python
# visualize
plt.scatter(data_tsne[:,0],data[:,1], c=labels,cmap='viridis')
plt.title("T-sne visualization of custmer segment")
plt.xlabel("t-sne component 1")
plt.ylabel("t-sne component 2")
plt.show()
```
